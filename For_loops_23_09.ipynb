```python
l = [4,5,12,21,12]
print(type(l))
print(len(l))
```

```python
l = [4,5,12,21,12]
l[::-1]
```

```python
l = [4,5,12,21,12]
for i in range(len(l)-1,-1,-1):
  print(l[i],end=" ")
```

```python
l = {4,5,12,21,12}
for i in range(len(l)-1,-1,-1):
  print(l[i],end=" ")
```

```python
l = {4,5,12,21,12}
k = set(sorted(l,reverse=True))
print(k)
```

```python
l = [4,5,12,21,12,2,3,3,4]
k = list(sorted(l,reverse=True))
print(k)
```

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot  as plt
```

```python
#histogram
df = pd.read_csv('churn_data.csv')
df
```

```python
df.info()
```

```python
df.groupby('PaymentMethod')['MonthlyCharges'].agg(['mean','median'])
```

```python
plt.figure(figsize = (6,5))
sns.barplot(data = df,x ='PaymentMethod',y='MonthlyCharges', errorbar = None)
plt.xlabel('PaymentMethod')
plt.ylabel('MonthlyCharges')
plt.title('Mean and median monthly payment distribution')
plt.xticks(rotation = 30)
plt.show()
```

```python
df_readian_monthly_charges = df.groupby(['PaymentMethod'])['MonthlyCharges'].median().reset_index().sort_values(by = 'MonthlyCharges', ascending = False)
df_readian_monthly_charges
```

```python
plt.figure(figsize = (6,5))
sns.barplot(data = df,x ='PaymentMethod',y='MonthlyCharges', errorbar = None,order = df_readian_monthly_charges['PaymentMethod'],estimator = 'median' )
plt.xlabel('PaymentMethod')
plt.ylabel('MonthlyCharges')
plt.title('Payment Method Distribution')
plt.xticks(rotation = 30)
plt.show()
```

```python
plt.figure(figsize = (6,5))
sns.boxplot(data = df,y='MonthlyCharges')
plt.xlabel('MonthlyCharges')
plt.ylabel('box plot')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('churn_data.csv')
df

"""Here are the counts of different payment method"""

df.info()

df.describe()

df.columns

payment_method_counts = df['PaymentMethod'].value_counts()
payment_method_counts

"""There is no pie chart method bulid in seaborn. Version = 0.13.2"""

plt.figure(figsize = (10,6))
plt.pie(x = payment_method_counts.values, labels = payment_method_counts.index, autopct= '%1.5f%%')
plt.title('Payment Method Distribution')
plt.show()
plt.pie

"""Add percentage"""

plt.figure(figsize = (10,6))
plt.pie(x = payment_method_counts.values, labels = payment_method_counts.index, autopct = '%.2f%%')
plt.title('Payment Method Distribution')
plt.show()

"""We could add annotations to show the percentage of each slice of the pie chart as well as explode a particular slice for emphasis:"""

plt.figure(figsize = (5, 3))
plt.pie(
    x = payment_method_counts.values,
    labels = payment_method_counts.index,
    autopct = '%.1f%%',
    explode = [0, 0, 0.1,0]
)
plt.title('Pie chart showing the distribution by payment method')
plt.show()

payment_method_counts

```
